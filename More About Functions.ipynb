### **Closure Function**

```python
def closure():
  msg = 'Hello'
  def display():
    print('*'*10)
    print(msg)
    print('*'*10)
  return display


##
d = closure()##Here, the function is returned
d()  ##Now we will call the Function okay.
```

```python
class Dept:
  def __init__(self):
    self.depts = {
                  'HR':'Human Resourse',
                  'Acc':'Accounts And Finance',
                  'SD':'Sales And Distribution',
                  'MKT':'Marketing'
                  }

  def __call__(self,dept):
    return self.depts[dept]


###Crete object
d = Dept()
s = d('HR')
print(s)
print(d('MKT'))


```

```python
class Depts:
    depts = {
        'HR': 'Human Resource',
        'Acc': 'Accounts And Finance',
        'SD': 'Sales And Distribution',
        'MKT': 'Marketing'
    }

    def dname(self, dept):
        return self.depts[dept]

# Create object
d = Depts()
s = d.dname('HR')
print(s)
print(d.dname('MKT'))
```

### **Decorator Function**

```python
def decorator(fun):
  def wrapper(msg):
    print('*'*10)
    fun(msg)
    print('*'*10)
  return wrapper


def display(msg):
  print(msg)

d = decorator(display)
d('Hello ')

```

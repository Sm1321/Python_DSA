### **Write a Python Program to Count the Frequenct of the words apperaing in a string**

ex:-


```python
def freq_words():
  str = input("Enter the String:-")
  lst = str.split()
  dic = {}
  for i in lst:
    if i not in dic.keys():
      dic[i] = 1  ##for the first time
    else:
      dic[i] = dic[i] + 1
  return dic


##Call the fun
freq_words()
```

### **Check the number is prime or NOT**

```python
def prime_check(num1):
  cnt = 0
  if num1 > 1:
    for i in range(2,num1-1):
      if (num1% 2 == 0):
        cnt+=1
    if cnt==0:
      return True
  return False


print(prime_check(7))

```

### **Find the Factorial of the number**

```python
def factorial(num1):
  ans = 1
  if num1 >= 1:
    for i in range(1, (num1+1) ):
      ans = i * ans
  else:
     return 0
  return ans

factorial(6)
```

### **Print the Fibonacci series**

```python
def fibanacci_series(num1,num2):
  a = num1
  b = num2
  print(a)
  print(b)
  for i in range(2,10):
    sum = a + b
    print(sum)
    a = b
    b = sum

fibanacci_series(0,1)
```

### **Decimal to Binary Conversion**

```python
def deci_to_bin(n):
  if n == 0:
    return '0'
  res = ""#Take the empty string
  while n > 0:
    res = res + str(n%2)
    n = n//2
  return res[::-1] #rev tha string


deci_to_bin(20)
```

### **Binary To DEcimal Conversion**

```python
## i/p :- b = 110
## o/p :- 6
##inbuilt function

##########
def bin_to_decimal(binary_str):
    # Convert the binary string to a decimal integer
    return int(binary_str, 2)

# Example usage:
b = "110"
result = bin_to_decimal(b)
print(result)  # Output will be 6



```

```python
def bin_to_decimal_1(n):
    res = 0
    p = 1
    # Convert the integer to a string to iterate over each digit
    for x in reversed(str(n)):
        res += int(x) * p
        p *= 2
    return res

# Example usage:
result = bin_to_decimal_1(1111)
print(result)  # Output will be 15
```

### **Get Smaller Elemnets**

```python
def smaller_ele_list(list1,num1):
  lst = []
  for ele in list1:
    if ele < num1:
      lst.append(ele)

  return lst


list1 = [10,20,30,40,1,2,3,5,4,7,8,9,11]
smaller_ele_list(list1,10)
```

### **Write a function  to Seperate Even and ODD from the list**

```python
def seperate_list(list1):
  lst1 = []
  lst2 = []
  for ele in list1:
    if ele % 2 == 0:
      lst1.append(ele)
    else:
      lst2.append(ele)
  return lst1,lst2

list1 = [1,0,2,4,5,63,9,8,7,4,2,3,6,69,1,36,6,67]
l1,l2 = seperate_list(list1)
print(l1)
print(l2)
```

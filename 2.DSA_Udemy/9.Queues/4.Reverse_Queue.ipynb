### First Approch :- Using Stack
- O(n) Time and O(n) Space

```python
from collections import deque

def reverseQueue(q):
    st=[]
    # Pop all the elements from the deque
    while q:
        st.append(q.popleft())
    # Push all the poped elemnsts from the stack  to the deque again     
    while st:
        q.append(st.pop())
            
        

q = deque()
q.append(1)
q.append(2)
q.append(3)

reverseQueue(q)

print(q)
```

### Second Approch :- Using Recursion

```python
from collections import deque

def revQ(q):
    if(len(q)==0):
        return
    x = q.popleft()
    revQ(q)
    q.append(x)
            
        

q = deque()
q.append(1)
q.append(2)
q.append(3)

revQ(q)

print(q)
```

```python
from collections import deque

class Solution:
    def generateNumbers(self, n):
        result = []
        q = deque()
        q.append("5")
        q.append("6")

        # Generate first n numbers
        while len(result) < n:
            curr = q.popleft()
            result.append(curr)
            # print(result)
            # print(q)

            # Generate next two numbers
            q.append(curr + "5")
            q.append(curr + "6")

        return result

# Example Usage
sol = Solution()
print(sol.generateNumbers(10))
```

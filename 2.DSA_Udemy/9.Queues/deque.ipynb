```python
from collections import deque

q = deque()
q.append(10)       # enqueue at right
q.append(20)
q.appendleft(5)    # add to left
print(q)           # deque([5, 10, 20])

print(q.popleft()) # removes 5 (front)
print(q.pop())     # removes 20 (end)
```

```python
from collections import deque

d = deque()

d.append(10)       # [10]
d.append(20)       # [10, 20]
d.appendleft(5)    # [5, 10, 20]

print(d.popleft()) # removes 5 → [10, 20]
print(d.pop())     # removes 20 → [10]

print(d)           # deque([10])
print(len(d))      # 1
```

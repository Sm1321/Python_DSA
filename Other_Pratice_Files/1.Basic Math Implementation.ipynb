### **Count the Digits in the Number**

```python
def count_Digits(num1):
  return len(str(num1))

num = int(input())
print("The lengt is :-",count_Digits(num))
```

### **Reverse a Number**

```python
def reverse_a_num(num1):
  ans = 0
  rem = 0
  while num1>0:
    rem = num1 % 10 ## thae the number and mul into 10
    ans = ans *10 + rem
    num1 = num1 // 10
  return "ans is:-",ans


print(reverse_a_num(4909525036))
```

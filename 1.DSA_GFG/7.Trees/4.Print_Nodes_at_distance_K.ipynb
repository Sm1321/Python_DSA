```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def height(node):
    if node is None:
        return 0
    LH = height(node.left)
    RH = height(node.right)
    return 1 + max(LH,RH )


root = Node(10)
root.left = Node(20)
root.right = Node(30)
root.left.left = Node(40)

```

## Print the Nodes at Distance K

```python
def PrintDistK(root,k):
    if root is None:
        return
    if k == 0:
        print(root.data,end = ",")
    #Call the left and Right  subtree
    else:
        PrintDistK(root.left,k-1)
        PrintDistK(root.right,k-1)


#Call the Function
PrintDistK(root,1)
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def height(node):
    if node is None:
        return 0
    LH = height(node.left)
    RH = height(node.right)
    return 1 + max(LH,RH )


root = Node(10)
root.left = Node(20)
root.right = Node(30)
root.left.left = Node(40)

```

```python
def postorder(node):
    if node is not None:
        postorder(node.left)
        postorder(node.right)
        print(node.data, end=" ")


def preorder(node):
    if node is not None:
        print(node.data, end=" ")
        preorder(node.left)
        preorder(node.right)



def inorder(node):
    if node is not None:
        inorder(node.left)
        print(node.data, end=" ")
        inorder(node.right)
```

```python
# Assuming S is the root node
print("Preorder:", end=" ")
preorder(root)
print("\nInorder:", end=" ")
inorder(root)
print("\nPostorder:", end=" ")
postorder(root)
print("\nHeight =", height(root))
```

## Print the Nodes at Distance K

```python
def PrintDistK(root,k):
    if root is None:
        return
    if k == 0:
        print(root.data,end = ",")
    #Call the left and Right  subtree
    else:
        PrintDistK(root.left,k-1)
        PrintDistK(root.right,k-1)


#Call the Function
PrintDistK(root,1)
```

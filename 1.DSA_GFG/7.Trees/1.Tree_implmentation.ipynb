```python
class Tree:
    def __init__(self,data):
        self.data = data
        self.left = None 
        self.right = None 

#Call the Tree
S = Tree(10)
S.left = Tree(20)
S.right = Tree(30)
S.right.left = Tree(40)
# S.right.right = Tree(42)     

```

```python
def PostOrder(node):
    if node is not None:
        PostOrder(node.left)     # visit left
        PostOrder(node.right)    # visit right
        print(node.data, end=" ")  # print root



#call the Inorder(Node):
PostOrder(S)      
```

```python
def PreOrder(node):
    if node is not None:
        print(node.data, end=" ")  # print root
        PreOrder(node.left)     # visit left
        PreOrder(node.right)    # visit right



#call the Inorder(Node):
PreOrder(S)      
```

```python
def Inorder(node):
    if node is not None:
        Inorder(node.left)     # visit left
        print(node.data, end=" ")  # print root
        Inorder(node.right)    # visit right



#call the Inorder(Node):
Inorder(S)      
```

```python

def postorder(node):
    if node is not None:
        postorder(node.left)
        postorder(node.right)
        print(node.data, end=" ")


def preorder(node):
    if node is not None:
        print(node.data, end=" ")
        preorder(node.left)
        preorder(node.right)



def inorder(node):
    if node is not None:
        inorder(node.left)
        print(node.data, end=" ")
        inorder(node.right)


# Assuming S is the root node

preorder(S)
print()
postorder(S)
print()
inorder(S)

```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def height(node):
    if node is None:
        return 0
    LH = height(node.left)
    RH = height(node.right)
    return 1 + max(LH,RH )


root = Node(10)
root.left = Node(20)
root.right = Node(30)
root.left.left = Node(40)

```

```python
def LevelOrderTraversal(root):
    if root is None:
        return
    #Call the left and Right  subtree
    print(root.data,end = ",")
    LevelOrderTraversal(root.left)
    LevelOrderTraversal(root.right)


#Call the Function
LevelOrderTraversal(root)
```
